```python
x['pixels']=pd.to_numeric(x['pixels'],errors='coerce')
x['pixels'].fillna(0,inplace=True)
x['pixels']=x['pixels']/255
age_dist=x['age'].value_counts()
ethnicity_dist = x['ethnicity'].value_counts()
gender_dist=x['gender'].value_counts().rename(index={0:'Male',1:'Female'})
def demo_bar(x, y, name):
    fig = go.Figure([go.Bar(x=x, y=y)])
    fig.update_layout(title_text=name)
    fig.show()
def demo_hist(data,name):
  fig=go.Figure([go.Histogram(x=data)])
  fig.update_layout(title_text=name)
  fig.show()
```
