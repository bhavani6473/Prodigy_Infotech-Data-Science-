```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
da=pd.read_csv('data set.csv')
da.head()
```

```python
da.drop(['Cabin'],axis=1,inplace=True)
da.head()
```

```python
da['Age'].fillna(da['Age'].median(),inplace=True)
da['Embarked'].fillna(da['Embarked'].mode()[0],inplace=True)
da.head()
```

```python
da['Age'].hist(bins=40,color='green',edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
x=da['Pclass'].value_counts()
plt.pie(x,labels=x.index,startangle=90, colors=['red', 'green', 'blue'])
plt.title("Passenger Class Distribution")
plt.show()
```

```python
import seaborn as s
x=da.corr()
s.heatmap(x,annot=True,cmap='BuPu')
```

```python
plt.fill_between(np.linspace(0,10,100),np.sin(np.linspace(0,10,100)),np.cos(np.linspace(0,10,100)),color='purple')
plt.title('Area Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
plt.scatter(np.random.rand(30),np.random.rand(30),color='orange')
plt.title("Scatterplot")
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.show()
```

```python
x=da.groupby('Age')['Fare'].sum().cumsum()
plt.plot(x.index,x.values,marker='o',linestyle='-',color='red')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.boxplot([da[da['Pclass'] == 1]['Age'],
             da[da['Pclass'] == 2]['Age'],
             da[da['Pclass'] == 3]['Age']],
            labels=['1st Class', '2nd Class', '3rd Class'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Age')
plt.ylabel('Passenger Class')
plt.show()
```
